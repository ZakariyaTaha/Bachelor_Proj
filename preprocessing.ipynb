```python
def transf(nb_pixels, arr):
  new_arr = np.zeros((arr.shape))
  new_arr[:, nb_pixels:len(arr[0]), :] = arr[:, 0:len(arr[0])-nb_pixels, :]
  new_arr[:,0:nb_pixels,:] = arr[:,len(arr[0])-nb_pixels:len(arr[0]),:]
  return new_arr.astype(int)
```
